# Summary Function
Pandas provide many simple "Summary Function" which restructure the data in some useful way.
- describe()
- mean()
- unique()
- value_counts()


```python
import pandas as pd
pd.set_option('display.max_rows',5)
winemag = pd.read_csv('winemag1.csv', index_col = 0)
winemag
```

```python
winemag.country.describe()
```

```python
winemag.price.describe()
```

```python
winemag.country.unique()
```

```python
winemag.country.value_counts()
```

```python
winemag.country.value_counts() == 1
```

```python
single_country = []
unique_value_counts = winemag['country'].value_counts()
for data, count in unique_value_counts.items():
    if count == 1:
        single_country.append(data)
single_country
```

# Map
A map is a term, borrowed from mathematics, for a function that takes one set of values and "maps" them to another set of values. In data science we often have a need for creating new representations from existing data, or for transforming data from the format it is in now to the format that we want it to be in later.

- map() => for series
- apply => for DataFrame

### Note:
- map() and apply() return new, transformed Series and DataFrames, respectively. They don't modify the original data they're called on.

```python
mean = winemag.points.mean()
winemag.points.map(lambda p: p - mean)
```

```python
def sub_mean(row):
    row.points = row.points - mean
    return row
winemag.apply(sub_mean, axis = 'columns')
```
