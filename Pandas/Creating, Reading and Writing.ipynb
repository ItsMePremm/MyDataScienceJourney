# Starting Pandas.

```python
import pandas as pd # This line help to use the function of Pandas library
print('Pandas is imported. You can Now Proceed to Further Task.')
```

## Creating Data

```python
# Using DataFrame
data = {'Name':['Prem', 'Pujan', 'Ram'], 
             'Age':[20, 10, 21],
             'Gender':['Male', 'Male', 'Male']}
pd.DataFrame(data)
```

```python
index_data = ['Present', 'Absent', 'Present']
pd.DataFrame(data, index = index_data) # index gives lable to the y-axis.
```

```python
pd.DataFrame(data, index = index_data, name = 'Info') # name is supported in Series but not in DataFrame
```

```python
# using Series
series_data = [30, 40, 50, 60]
pd.Series(series_data)
```

```python
series_index = [2017, 2018, 2019, 2020]
pd.Series(series_data, index = series_index, name = 'Sale in Four Year')
```
