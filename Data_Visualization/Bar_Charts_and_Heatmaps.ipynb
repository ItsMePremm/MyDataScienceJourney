```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup Successful')
```

```python
pd.read_csv('flight_delays.csv')
```

```python
delay_data = pd.read_csv('flight_delays.csv', index_col = 'Month')
```

```python
delay_data.head()
```

```python
plt.figure(figsize = (10,6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')
sns.barplot(x = delay_data.index, y = delay_data['US'])
plt.ylabel('Arrival delay (in minutes)')
```

```python
plt.figure(figsize = (10,6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')
sns.barplot(x = delay_data.index, y = delay_data['NK'])
plt.ylabel('Arrival delay (in minutes)')
```

```python
plt.figure(figsize = (20,12))
plt.title('Heatmap for Average Arrival Delay For Each Airline, by Month')
sns.heatmap(data = delay_data, annot = True)
plt.xlabel('Airline')
```

```python
plt.figure(figsize = (20,12))
plt.title('Heatmap for Average Arrival Delay For Each Airline, by Month')
sns.heatmap(data = delay_data)
plt.xlabel('Airline')
```
